```python
import numpy as np
from sklearn import datasets

import pandas as pd
from pandas import DataFrame as df

import matplotlib.pyplot as plt
```

### Loading of the Iris dataset


```python
#Load iris dataset

iris = datasets.load_iris()
X = iris.data
Y = iris.target

#Combining features and target to see the whole data altogether.

X = np.insert(X, X.shape[1], Y,  axis=1)
data = pd.DataFrame(X)

data = data.reindex(np.random.permutation(data.index))
data.head()
```

### Assignig Class Labels and viewing the target class

```python
features = data.iloc[:,:4]
target = data.iloc[:,4]

target = target.values.reshape((target.shape[0],1)) # Reshaping target for matrix multiplication
target_class = np.zeros(shape=(target.shape[0],3))  # Initializing class labels as [0,0,0]

# Assigning class labels as [1,0,0], [0,1,0] and [0,0,1]

for i in range(target.shape[0]):
    if target[i] == 0:
        target_class[i][0] = 1
    elif target[i]==1:
        target_class[i][1] = 1
    else:
        target_class[i][2] = 1

print(target_class.shape)
print('\n')
print(target_class)
```

### Definition of the Sigmoid Activation function

```python
#Defining activation function (here sigmoid)

def sigmoid(x):
    return 1/(1+np.exp(-x))

#Defining derivative of sigmoid function

def der_sigmoid(x):
    return(sigmoid(x)*(1-sigmoid(x)))
```

### Defining the layer-wise node-count of the Single Layer NN

```python
input_nodes = features.shape[1]
output_nodes = 3

#Initializing weight matrix of size input_nodes x output_nodes with random values.
weight_matrix = np.random.uniform(size=(input_nodes, output_nodes))

#Assigning epoch and learning rate value.
epoch = 1000
lr = 0.005
```

### Weight Updation

```python
#Updating weights

for i in range(epoch):
    output_node_input = features.dot(weight_matrix)
    output = sigmoid(output_node_input)
    
    error = target_class - output
    drv = der_sigmoid(output_node_input)
    delta_weight = error*drv
    
    weight_matrix = weight_matrix + lr*features.T.dot(delta_weight)
    
```

### Making the Predictions

```python
final_output = np.array(output)
print('Final Output:\n')
print(final_output, '\n')

output_class = np.zeros(shape=final_output.shape[0])

for i in range(final_output.shape[0]):
    output_class[i] = np.argmax(final_output[i])

#print('Output_class:\n') This has been done in the next cell for better understandability
#print(output_class)

class_diff = output_class - target[:,0]
```

### Comparison of Training Examples with Predicted Results

```python
print('Actual Class')
print(target[:,0])

print("\n>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>>\n")

print('Predicted Class')
print(output_class)

wrong_prediction = np.count_nonzero(class_diff)

N = len(class_diff)

simple_accuracy = 100 * (N-wrong_prediction)/N
print("\nAccuracy : ",(simple_accuracy) ,'%')
```
