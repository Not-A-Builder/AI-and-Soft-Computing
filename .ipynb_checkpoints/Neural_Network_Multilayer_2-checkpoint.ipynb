# Constructing a Multi-layer Neural Network - Example 2

In order to enable higher accuracy, we need to give more freedom to the neural network.
This means that a neural network needs more than one layer to extract the underlying
patterns in the training data.

```python
import numpy as np
import matplotlib.pyplot as plt
import neurolab as nl
%matplotlib inline
```

In the previous two examples, we saw how to use a neural network as a <b>classifier</b>.<br/> In this
section, we will see how to use a multilayer neural network as a <b>regressor</b>.

<h5>Generate some sample data points based on the equation y = 3x^2 + 5 and then normalize the points.</h5>

```python
min_val = -15
max_val = 15
num_points = 130
#linspace(start of interval range, end of interval range, No. of samples to generate)
x = np.linspace(min_val, max_val, num_points)
y = 3 * np.square(x) + 5
y /= np.linalg.norm(y)
```

```python
# Reshape the above variables to create a training dataset
data = x.reshape(num_points, 1)
labels = y.reshape(num_points, 1)
```

```python
# Plot input data
plt.figure()
plt.scatter(data, labels)
plt.xlabel('Dimension 1')
plt.ylabel('Dimension 2')
plt.title('Input data')
```

<h5>Define a multilayer neural network with 2 hidden layers. <br/>You are free to design a neural network any way you want. <br/>For this case, let's have 10 neurons in the 1st layer and 6 neurons in the 2nd layer. <br/>Our task is to predict the value, so the output layer will contain a single neuron.</h5>

```python
# Create a feed-forward back propagation network
# 1st argument is min max values of predictor variables; can be a list of list
# 2nd argument is no.of nodes in each layer i.e 16 in hidden & 1 in output layer
nn = nl.net.newff([[min_val, max_val]], [10, 6, 1])
```

```python
# Set the training algorithm to gradient descent
nn.trainf = nl.train.train_gd
```

```python
# Train the neural network using the training data that was generated.
error_progress = nn.train(data, labels, epochs=2000, show=100, goal=0.01)
```

```python
# Run the neural network on the training datapoints.
output = nn.sim(data)
y_pred = output.reshape(num_points)
```

```python
# Plot the training progress
plt.figure()
plt.plot(error_progress)
plt.xlabel('Number of epochs')
plt.ylabel('Error')
plt.title('Training error progress')
```

```python
# Plot the predicted output
x_dense = np.linspace(min_val, max_val, num_points * 2)
y_dense_pred = nn.sim(x_dense.reshape(x_dense.size,1)).reshape(x_dense.size)

plt.figure()
plt.plot(x_dense, y_dense_pred, '-', x, y, '.', x, y_pred, 'p')
plt.title('Actual vs predicted')

plt.show()
```

The predicted output seems to follow the general trend. If you continue to train the network
and reduce the error, you will see that the predicted output will match the input curve even
more accurately.
